```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def show_with_plt(color_img, title):

    #convert BGR to RGB
    img_RGB = color_img[:, :, ::-1]

    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis("off")

plt.figure(figsize=(15, 8))
plt.suptitle("Arithmetic with Image", fontsize=14, fontweight="bold")
```

```python
image = cv2.imread("images/lenna.png")
show_with_plt(image, "Original Image")
```

###### Adding 60 to every pixel in the image

```python
M = np.ones(image.shape, dtype="uint8") * 60
added_image = cv2.add(image, M)
show_with_plt(added_image, "Added 60 to image pixels")
```

###### Substracting 60 from every pixel

```python
substracted_image = cv2.subtract(image, M)
show_with_plt(substracted_image, "Subtracted Image")
```

```python
# Additionally, we can build an scalar and add/subtract it


scalar = np.ones((1, 3), dtype="float") * 110
added_image_2 = cv2.add(image, scalar)
show_with_plt(added_image_2, "Added 110 (Image + Scalar)")
```

```python
sub_image_2 = cv2.subtract(image, scalar)
show_with_plt(sub_image_2, "Subtracted 110 (Image -Scalar)")
```
