```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def show_with_plt(color_img, title):
    
    #convert BGR to RGB
    img_RGB = color_img[:, :, ::-1]

    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis("off")

plt.figure(figsize=(12, 6))
plt.suptitle("Comparing Different Kernels using cv2.filter2D()", fontsize=14, fontweight="bold")
```

```python
image = cv2.imread("images/cat-face.png")
```

```python
#IDENTITY kernel does not modify the image 
kernel_identity = np.array([[0,0,0],
                            [0,1,0],
                            [0,0,0]])
```

###### Kernels for EDGE Detection

```python
kernel_edge_1 = np.array([[1, 0, -1],
                          [0,0,0],
                          [-1, 0, 1]])

kernel_edge_2 = np.array([[0,1,0],
                          [1,-4,1],
                          [0,1,0]])

kernel_edge_3 = np.array([[-1,-1,-1],
                          [-1, 8,-1],
                          [-1,-1,-1]])
```

###### kernels for sharpening

```python
kernel_sharpen = np.array([[0,-1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

kernel_unsharp_masking = -1 / 256 * np.array([[1, 4, 6, 4, 1],
                                              [4,16, 24,16,4],
                                              [6,24,-476,24,6],
                                              [4,16,24,16,4],
                                              [1,4,6,4,1]])
```

###### Kernel for SMOOTHING

```python
kernel_blur = 1/9 * np.array([[1, 1, 1],
                              [1, 1, 1],
                              [1, 1, 1]])

gaussian_blur = 1 /16 * np.array([[1, 2, 1],
                                  [2, 4, 2],
                                  [1, 2, 1]])
```

###### Kernel for EMBOSSING

```python
kernel_emboss = np.array([[-2, -1, 0],
                          [-1, 1, 1],
                          [0, 1, 2]])
```

###### SOBEL kernels for EDGE Detection

```python
sobel_x_kernel = np.array([[1, 0, -1],
                           [2, 0, -2],
                           [1, 0, -1]])

sobel_y_kernel = np.array([[1, 2, 1],
                           [0,0,0],
                           [-1, -2, -1]])

outline_kernel = np.array([[-1, -1, -1],
                           [-1, 8, -1],
                           [-1, -1, -1]])
```

##### Applying all the KERNELS

```python
original_image = cv2.filter2D(image, -1, kernel_identity)
show_with_plt(original_image, "Identity Kernel")
```

```python
edge_image_1 = cv2.filter2D(image, -1, kernel_edge_1)
show_with_plt(edge_image_1, "EDGE Detection 1")
```

```python
edge_image_2 = cv2.filter2D(image, -1, kernel_edge_2)
show_with_plt(edge_image_2, "EDGE Detection 2")
```

```python
edge_image_3 = cv2.filter2D(image, -1, kernel_edge_3)
show_with_plt(edge_image_3, "EDGE Detection 3")
```

```python
sharpen_image = cv2.filter2D(image, -1, kernel_sharpen)
show_with_plt(sharpen_image, "Sharpen Image")
```

```python
unsharp_masking_image = cv2.filter2D(image, -1, kernel_unsharp_masking)
show_with_plt(unsharp_masking_image, "Unsharp Masking")
```

```python
blur_image = cv2.filter2D(image, -1, kernel_blur)
show_with_plt(blur_image, "Blurred Image")
```

```python
gaussian_blur_image = cv2.filter2D(image, -1,gaussian_blur)
show_with_plt(gaussian_blur_image, "Guassian Blurred Image")
```

```python
embos_image = cv2.filter2D(image, -1,kernel_emboss)
show_with_plt(embos_image, "Emboss Image")
```

```python
sobel_x_image = cv2.filter2D(image, -1,sobel_x_kernel)
show_with_plt(sobel_x_image, "Sobel X-Image")
```

```python
sobel_y_image = cv2.filter2D(image, -1,sobel_y_kernel)
show_with_plt(sobel_y_image, "Sobel Y-Image")
```

```python
outline_image = cv2.filter2D(image, -1,outline_kernel)
show_with_plt(outline_image, "Outline Image")
```
