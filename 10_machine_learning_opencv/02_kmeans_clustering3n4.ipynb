##### Clustering for 3 i.e ``K=3``

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

```python
#creating data 
data3 = np.float32(np.vstack(
    (np.random.randint(0, 40, (50, 2)), np.random.randint(30, 70, (50, 2)), np.random.randint(60, 100, (50, 2)))
))
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 1.0)
```

```python
#Applying K-MEANS with k=3, and attempts =10

ret, label, center = cv2.kmeans(data3, 3, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
#Now separating the data using label output (sores the cluster indices for every sample)
#Therefore, we split the data to different clusters depending on their labels 

A = data3[label.ravel() == 0]
B = data3[label.ravel() == 1]
C = data3[label.ravel() == 2]
```

```python
fig = plt.figure(figsize=(12, 8))
plt.rcParams['figure.facecolor'] = 'silver'
```

```python
#Plotting the ORIGINAL DATA
ax = plt.subplot(1,2, 1)
plt.scatter(data3[:, 0], data3[:, 1], c ='c')
plt.title("Original Data")


#Plotting the clustered data
ax=plt.subplot(1, 2, 2)
plt.scatter(A[:, 0], A[:, 1], c='b')
plt.scatter(B[:, 0], B[:, 1], c='g')
plt.scatter(C[:, 0], C[:, 1], c='r')
plt.scatter(center[:, 0], center[:, 1], s=100, c='m',marker='s')
plt.title("Clustered Data and Centroid (K -3)")

plt.subplots_adjust(top=3,bottom=1.5, right=2, left=0.2, wspace=0.3, hspace=0.3)
plt.show()
```

#### Clustering for 4 ``(K=4)``

- data will be the ``data3``
- criteria will be the same ``criteria``

```python
#Applying K-MEANS for K=4, attempts=10

ret2, label2, center2 = cv2.kmeans(data3, 4, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
#Now separting the data using label 

A = data3[label2.ravel() == 0]
B = data3[label2.ravel() == 1]
C = data3[label2.ravel() == 2]
D = data3[label2.ravel() == 3]
```

```python
#Plotting the ORIGINAL DATA
ax = plt.subplot(1,2, 1)
plt.scatter(data3[:, 0], data3[:, 1], c ='c')
plt.title("Original Data")


#Plotting the clustered data
ax=plt.subplot(1, 2, 2)
plt.scatter(A[:, 0], A[:, 1], c='b')
plt.scatter(B[:, 0], B[:, 1], c='g')
plt.scatter(C[:, 0], C[:, 1], c='r')
plt.scatter(D[:, 0], D[:, 1], c='y')
plt.scatter(center[:, 0], center[:, 1], s=100, c='m',marker='s')
plt.title("Clustered Data and Centroid (K =4)")

plt.subplots_adjust(top=3,bottom=1.5, right=2, left=0.2, wspace=0.3, hspace=0.3)
plt.show()
```
