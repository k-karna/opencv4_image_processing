```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
```

```python
def build_sample_image():

    #defining different tones
    tones = np.arange(start=50, stop=300, step=50)

    #initialize result with the first 50x50 region with 0-intensity level
    result = np.zeros((50, 50, 3), dtype="uint8")

    for tone in tones:
        img = np.ones((50, 50, 3), dtype="uint8") * tone 
        result = np.concatenate((result, img), axis=1)

    return result 
```

```python
def img_plt(color_img, title, pos):

    img_RGB = color_img[:, :, ::-1]

    ax = plt.subplot(8, 1, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis("off")
```

```python
fig = plt.figure(figsize=(5, 9))
plt.suptitle("Simple Thresholding Types", fontsize=14)
plt.rcParams['figure.facecolor'] = 'silver'
```

```python
image = build_sample_image()
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

Applying different __THRESHOLD TYPES__

```python
ret1, thresh1 = cv2.threshold(gray_image, 100, 255, cv2.THRESH_BINARY)
ret2, thresh2 = cv2.threshold(gray_image, 100, 220, cv2.THRESH_BINARY)

ret3, thresh3 = cv2.threshold(gray_image, 100, 255, cv2.THRESH_BINARY_INV)
ret4, thresh4 = cv2.threshold(gray_image, 100, 220, cv2.THRESH_BINARY_INV)

ret5, thresh5 = cv2.threshold(gray_image, 100, 255, cv2.THRESH_TRUNC)
ret6, thresh6 = cv2.threshold(gray_image, 100, 255, cv2.THRESH_TOZERO)
ret7, thresh7 = cv2.threshold(gray_image, 100, 255, cv2.THRESH_TOZERO_INV)
```

Plotting

```python
img_plt(cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR), "IMG TONES 0 --> 255",1)

img_plt(cv2.cvtColor(thresh1, cv2.COLOR_GRAY2BGR), "THRESH_BINARY with MaxValue 255", 2)
img_plt(cv2.cvtColor(thresh2, cv2.COLOR_GRAY2BGR), "THRESH_BINARY with MaxValue 220", 3)
img_plt(cv2.cvtColor(thresh3, cv2.COLOR_GRAY2BGR), "THRESH_BINARY_INV", 4)
img_plt(cv2.cvtColor(thresh4, cv2.COLOR_GRAY2BGR), "THRESH_BINARY_INV with MaxVALUE 220",5)
img_plt(cv2.cvtColor(thresh5, cv2.COLOR_GRAY2BGR), "THRESH_TRUNC",6)
img_plt(cv2.cvtColor(thresh6, cv2.COLOR_GRAY2BGR), "THRESH_TOZERO",7)
img_plt(cv2.cvtColor(thresh7, cv2.COLOR_GRAY2BGR), "THRESH_TOZERO_INV",8)

plt.subplots_adjust(top=1.5,bottom=0.1, right=2, left=1, wspace=1, hspace=1)
```
