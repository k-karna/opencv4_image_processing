```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
def img_plt(color_img, title, pos):

    img_RGB = color_img[:, :, ::-1]
    ax = plt.subplot(5, 2, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis("off")
```

```python
fig = plt.figure(figsize=(20, 20))
plt.suptitle("Thresholding Sudoku", fontsize=14)
plt.rcParams['figure.facecolor'] = 'silver'
```

```python
image = cv2.imread("images/sudoku.png")
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

Applying THRESHOLD with different values

```python
ret1, thresh1 = cv2.threshold(gray_image, 60, 255, cv2.THRESH_BINARY)
ret2, thresh2 = cv2.threshold(gray_image, 70, 255, cv2.THRESH_BINARY)
ret3, thresh3 = cv2.threshold(gray_image, 80, 255, cv2.THRESH_BINARY)
ret4, thresh4 = cv2.threshold(gray_image, 90, 255, cv2.THRESH_BINARY)
ret5, thresh5 = cv2.threshold(gray_image, 100, 255, cv2.THRESH_BINARY)
ret6, thresh6 = cv2.threshold(gray_image, 110, 255, cv2.THRESH_BINARY)
ret7, thresh7 = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY)
ret8, thresh8 = cv2.threshold(gray_image, 130, 255, cv2.THRESH_BINARY)
```

Plotting first the original Image, and its grayscale and then 8 different with THRESHOLD Values

```python
img_plt(image, "Original Image", 1)
img_plt(cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR), "GrayScale Image", 2)

img_plt(cv2.cvtColor(thresh1, cv2.COLOR_GRAY2BGR), "Threshold Value: 60", 3)
img_plt(cv2.cvtColor(thresh2, cv2.COLOR_GRAY2BGR), "Threshold Value: 70", 4)
img_plt(cv2.cvtColor(thresh3, cv2.COLOR_GRAY2BGR), "Threshold Value: 80", 5)
img_plt(cv2.cvtColor(thresh4, cv2.COLOR_GRAY2BGR), "Threshold Value: 90", 6)
img_plt(cv2.cvtColor(thresh5, cv2.COLOR_GRAY2BGR), "Threshold Value: 100",7)
img_plt(cv2.cvtColor(thresh6, cv2.COLOR_GRAY2BGR), "Threshold Value: 110", 8)
img_plt(cv2.cvtColor(thresh7, cv2.COLOR_GRAY2BGR), "Threshold Value: 120", 9)
img_plt(cv2.cvtColor(thresh8, cv2.COLOR_GRAY2BGR), "Threshold Value: 130", 10)

plt.subplots_adjust(top=2.8,bottom=0.3, right=0.8, left=0.1, wspace=0.5, hspace=0.5)
plt.show()
```
