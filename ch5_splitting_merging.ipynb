```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
def show_with_plt(color_img, title, pos):

    #convert BGR image to RGB
    img_RGB = color_img[:,:,::-1]

    ax = plt.subplot(3, 6, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis("off")
```

```python
#loading the original image 

image = cv2.imread("images/color_spaces.png")
```

```python
#creating a figure object with appropriate size and title 

plt.figure(figsize=(15, 5))
plt.suptitle("Splitting and Merging Channels", fontsize=14, fontweight="bold")

#show the BGR image
show_with_plt(image, "BGR - image",1)
```

###### Splitting the three channels of the image

```python
(b, g, r) = cv2.split(image)
```

```python
#show all the channels from the BGR image 

show_with_plt(cv2.cvtColor(b, cv2.COLOR_GRAY2BGR), "BGR - (B)", 2)
show_with_plt(cv2.cvtColor(g, cv2.COLOR_GRAY2BGR), "BGR - (G)", 2+6)
show_with_plt(cv2.cvtColor(r, cv2.COLOR_GRAY2BGR), "BGR - (R)", 2+6*2)
```

###### Merging all the three channels again to build BGR image

```python
image_copy = cv2.merge((b, g, r))

show_with_plt(image_copy, "BGR IMAGE copy", 1+6)
```

``cv2.split`` is a time consuming operation. Therefore, Numpy is more efficient for separating channels

```python
image_without_blue = image.copy()
image_without_green = image.copy()
image_without_red = image.copy()
```

```python
image_without_blue[:, :, 0] = 0 #setting blue to 0
image_without_green[:, :, 1] = 0 #setting green to 0
image_without_red[:, :, 2] = 0 #setting red to 0
```

```python
show_with_plt(image_without_blue, "BGR with BlUE Channel",3)
show_with_plt(image_without_green, "BGR with GREEN Channel",3+6)
show_with_plt(image_without_red, "BGR with RED Channel",3+6*2)
```

#### Testing Section

```python
(b, g, r) = cv2.split(image_without_blue)

show_with_plt(cv2.cvtColor(b, cv2.COLOR_GRAY2BGR), "BGR without Blue (Blue Channel)", 4)
show_with_plt(cv2.cvtColor(g, cv2.COLOR_GRAY2BGR), "BGR without Blue (Green Channel)", 4+6)
show_with_plt(cv2.cvtColor(r, cv2.COLOR_GRAY2BGR), "BGR without Blue (Red Channel)", 4+6*2)
```

```python
(b, g, r) = cv2.split(image_without_green)

show_with_plt(cv2.cvtColor(b, cv2.COLOR_GRAY2BGR), "BGR without Green (Blue Channel)", 4)
show_with_plt(cv2.cvtColor(g, cv2.COLOR_GRAY2BGR), "BGR without Green (Green Channel)", 4+6)
show_with_plt(cv2.cvtColor(r, cv2.COLOR_GRAY2BGR), "BGR without Green (Red Channel)", 4+6*2)
```
